```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Calculus'
req = requests.get(url)
webpage = req.text

soup = BeautifulSoup(webpage, 'html.parser')
paragraphs = soup.find_all('p')

paragraphs[1]
```

```python
import pandas as pd

dic = {'class': [], 'href': [], 'title': []}

for paragraph in paragraphs:
    for entry in paragraph.find_all('a', attrs={'class':True, 'title':True}):
        dic['class'].append(entry['class'])
        dic['href'].append(entry['href'])
        dic['title'].append(entry['title'])
        
df = pd.DataFrame(dic)
df
```

```python
df.to_csv('Sample_data.csv')
```
